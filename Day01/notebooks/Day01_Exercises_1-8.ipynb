## Exercise 1 — Your First HTML Page (EX01)

### What you learn
Every HTML file has a **skeleton**: doctype, `<html>`, `<head>`, and `<body>`. The browser reads this structure to show your page.

### සිංහලෙන් (Sinhala)
HTML ගොනුවක **මූලික ව්‍යුහයක් (skeleton)** තිබෙනවා. `<head>` තුළ title වැනි තොරතුරු තියෙනවා. `<body>` තුළ පරිශීලකයාට පෙනෙන අන්තර්ගතය (ශීර්ෂය, ඡේද) තියෙනවා.

### 8 Steps
1. Start with `<!DOCTYPE html>` — tells the browser this is HTML5.
2. Open `<html lang="en">` — the whole page lives inside this.
3. Open `<head>` — information for the browser, not visible on the page.
4. Add `<meta charset="UTF-8">` — supports letters like Sinhala, Tamil, emoji.
5. Add `<meta name="viewport" ...>` — helps the page look good on phones.
6. Add `<title>...</title>` — text on the browser tab.
7. Open `<body>` and add `<h1>` (big heading) and `<p>` (paragraph).
8. Close all tags (`</body>`, `</html>`) and save as `.html`, then open in a browser.

**සිංහල:** පියවර 1–7 එකින් එක type කරන්න. පියවර 8 — `.html` ලෙස save කර browser එකෙන් open කරන්න.

```python
from IPython.display import HTML, display

ex01_html = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>This is the title</title>
</head>
<body>
    <h1>This is a heading</h1>
    <p>Document content goes here...</p>
</body>
</html>
"""

print("--- HTML code for EX01 ---")
print(ex01_html)
print("\n--- Preview in notebook ---")
display(HTML(ex01_html))
```

## Exercise 2 — Six Heading Levels (EX02)

### What you learn

### 8 Steps

5. Continue with `<h3>`, `<h4>`, `<h5>`.
6. Add `<h6>` — the smallest heading.
7. Use only **one** `<h1>` per page when possible (best practice).
8. Run the preview cell below and compare sizes.

**සිංහල:** h1 සිට h6 දක්වා එකින් එක add කරලා size වෙනස්කම බලන්න.

## Exercise 3 — Paragraphs (EX03)

### What you learn
A **paragraph** is a block of text. Each `<p>...</p>` starts on a new line with space above and below.

### සිංහලෙන්
`<p>` tag එකක් භාවිතා කර text block එකක් ලියනවා. paragraph එකක් අවසන් වෙලා ඊළඟ `<p>` නව පේළියකින් පටන් ගනී.

### 8 Steps
1. Start with the HTML skeleton (doctype, html, head, body).
2. Set a useful `<title>` (e.g. "Paragraph Example").
3. In `<body>`, write your first paragraph: `<p>First paragraph.</p>`.
4. Add a second `<p>` with different text.
5. Add a third `<p>`.
6. Notice: you do **not** press Enter inside `<p>` for a new paragraph — use a **new** `<p>` tag.
7. Preview in the browser or notebook.
8. Try changing the text to something about yourself (name, school).

**සිංහල:** එක් ඡේදයක් = එක් `<p>` tag එක. Enter key එකෙන් අලුත් ඡේදයක් ආරම්භ කළ නොහැක — නව `<p>` භාවිතා කරන්න.

```python
ex03_html = """
<p>This is a first paragraph of text.</p>
<p>This is a second paragraph of text.</p>
<p>This is a third paragraph of text.</p>
"""
display(HTML(ex03_html))
```

## Exercise 4 — Line Breaks with `<br>` (EX04)

### What you learn
Sometimes you want a **new line inside the same paragraph** (like a short letter). Use `<br>` — it does not need a closing tag.

### සිංහලෙන්
එකම paragraph එක තුළ නව පේළියක් ඕන නම් `<br>` භාවිතා කරන්න. උදා: ලිපියක්, ලිපිනයක්.

### 8 Steps
1. Create the basic HTML page structure.
2. Open one `<p>` tag.
3. Type the first line of text (e.g. "Hello").
4. Add `<br>` to break to the next line.
5. Type the second line.
6. Add another `<br>` and continue for more lines.
7. Close with `</p>`.
8. Compare: many lines in **one** `<p>` + `<br>` vs **many** `<p>` tags (different spacing).

**සිංහල:** `<br>` = line break. closing tag එකක් නැහැ (`<br>` පමණක්).

```python
ex04_html = """
<p>Hello<br>You delivered your assignment on time.<br>Thanks<br>Mahnaz</p>
"""
display(HTML(ex04_html))
```

## Exercise 5 — Text Formatting (EX05)

### What you learn
HTML can change how text **looks** and what it **means**:
- `<b>` — bold (visual)
- `<strong>` / `<em>` — important (meaning)
- `<i>` — italic
- `<small>` — smaller text
- `<sub>` — subscript (e.g. H₂O)
- `<sup>` — superscript (e.g. x²)

### සිංහලෙන්
- `<b>` — තද අකුරු (bold)
- `<em>` — වැදගත් (emphasis)
- `<i>` — ඇල අකුරු (italic)
- `<small>` — කුඩා අකුරු
- `<sub>` — පහළ ලියන අකුරු | `<sup>` — ඉහළ ලියන අකුරු

### 8 Steps
1. Set up your HTML page with `<body>`.
2. Add a normal paragraph: `<p>This text is Normal</p>`.
3. Wrap text in `<b>...</b>` for bold.
4. Put `<b>` only around one word inside a sentence.
5. Try `<em>` for important text.
6. Try `<i>` for italic style.
7. Add `<small>`, then `<sub>` and `<sup>` in one line.
8. Preview and say which tags change **look** vs **meaning**.

**සිංහල:** bold පෙනුමට `<b>`, වැදගත්කමට `<em>` — දෙකම භාවිතා කරන්න පුළුවන්.

```python
ex05_html = """
<p>This text is Normal</p>
<p><b>This text is BOLD</b></p>
<p>This text is also <b>BOLD</b></p>
<p><em>This text is important!</em></p>
<p><i>This text is Italic</i></p>
<p><small>This text is small</small></p>
<p>This is a <sub>subscript</sub> and this is a <sup>superscript</sup></p>
"""
display(HTML(ex05_html))
```

## Exercise 6 — CSS Styling (EX06)

### What you learn
**CSS** controls colors, fonts, size, and alignment. You can write CSS:
- **Inside `<style>` in `<head>`** — affects the whole page (e.g. `body { background-color: yellow; }`)
- **Inline on a tag** — `style="color:blue;"` on one element only

### සිංහලෙන්
CSS මගින් පිටුවේ වර්ණ, අකුරු, ප්‍රමාණය, alignment වෙනස් කරනවා.
- `<style>` — මුළු පිටුවට
- `style="..."` — එක් element එකට පමණක්

### 8 Steps
1. Add `<style>` inside `<head>`.
2. Set `body { background-color: yellow; }` for the page background.
3. Add an `<h1>` in the body.
4. Use inline style: `style="font-family:verdana"` on a paragraph.
5. Color text: `style="color:blue;"`.
6. Center and resize: `style="text-align:center; font-size:50px;"`.
7. Align left on another paragraph: `text-align:left`.
8. Try `<mark>` with custom background and text color.

**සිංහල:** `background-color`, `color`, `font-size`, `text-align` — මේ properties ප්‍රයෝගිකව වෙනස් කර බලන්න.

```python
ex06_html = """
<style>
    body { background-color: yellow; }
</style>
<h1>This is my Heading</h1>
<p style="font-family:verdana">This is a paragraph.</p>
<p>I am normal</p>
<mark style="background-color:green; color:red; display:block;">This font color is red</mark>
<p style="color:blue;">This font color is blue</p>
<p style="text-align:center; font-size:50px;">This text is centered and large</p>
<p style="text-align:left;">I'm align left</p>
"""
display(HTML(ex06_html))
```

## Exercise 7 — Links, Images, and Buttons (EX07)

### What you learn
- `<a href="URL">text</a>` — **link** (click to go somewhere)
- `<img src="file.jpg" alt="description">` — **image**
- `<button>` inside `<a>` — looks like a button but acts as a link

### සිංහලෙන්
- `<a href="...">` — සබැඳිය (link). `#` = තාවකාලික (placeholder)
- `<img src="..." alt="...">` — රූපය. `alt` = රූපය load නොවුණොත් පෙනෙන වචන
- `<button>` — බොත්තම; `<a>` ඇතුළත දාලා link ලෙසත් භාවිතා කරන්න පුළුවන්

### 8 Steps
1. Create your page with `<h1>My Web</h1>`.
2. Write a paragraph about yourself.
3. Wrap a word (e.g. school name) in `<a href="#">SEUSL</a>`.
4. Add `<h2>` for a new section (e.g. "Image of SEUSL").
5. Add `<img src="seusl.jpg" alt="SEUSL Image" style="width:100%;">`.
6. Put image file in the **same folder** as your HTML (or fix the path).
7. Add `<h2>Button as a link</h2>` and wrap `<button>` with `<a href="#">`.
8. Open in browser; click the link and button to test.

**සිංහල:** `seusl.jpg` ගොනුව HTML ගොනුව තිබෙන folder එකේ තිබිය යුතුයි.

```python
ex07_html = """
<h1>My Web</h1>
<p>I am Nethsara. I am a second-year student at <a href="#">SEUSL</a>. I live in Nuwaraeliya.</p>
<h2>Image of SEUSL</h2>
<p><em>(Image shows only if ../assets/seusl.jpg is in the same folder as this notebook/HTML file.)</em></p>
<img src="../assets/seusl.jpg" alt="SEUSL Image" style="width:100%; height:auto;">
<h2>Button as a link</h2>
<a href="#"><button type="button">FAS page</button></a>
"""
display(HTML(ex07_html))
```

## Exercise 8 — Full Page: My Cat (EX08)

### What you learn
Combine everything: **structure** (headings, paragraphs, list), **CSS** (colors, fonts, float image), and **links**. This is a small real webpage.

### සිංහලෙන්
මෙහි ඔබ ඉගෙන ගත් දේවල් එකට එකතු කරනවා:
- ශීර්ෂ (h1, h2), ඡේද (p), අනුපිළි (ol, li)
- CSS: පසුබිම් වර්ණ, අකුරු, `<strong>` රතු වර්ණය
- රූපය දකුණට: `float: right`
- සබැඳි (links)

### 8 Steps
1. Add `<style>` in `<head>`: light blue background, padding on `body`.
2. Style `h1, h2` with `font-family: monospace`.
3. Make `<strong>` red for highlighted words (e.g. city name).
4. Write Introduction: heading + paragraph with `<strong>Seattle, WA</strong>`.
5. Add Profile section: `<ol>` with three `<li>` items (use `<i>` for labels).
6. Add Links section with two `<a href="#">` links and `<br>` between them.
7. Add image: `<img src="cat.jpg" class="cat-image">` and CSS `.cat-image { float: right; width: 200px; }`.
8. Save as `EX08.html`, add `cat.jpg` to the folder, open in browser and check layout.

**සිංහල:** මෙය Day 01 හි සම්පූර්ණ අභ්‍යාසය — structure + style + content එකට.

```python
ex08_html = """
<style>
    body {
        background-color: lightblue;
        font-family: serif;
        padding: 20px;
    }
    h1, h2 { font-family: monospace; }
    strong { color: red; }
    .cat-image {
        float: right;
        width: 200px;
        margin-top: -50px;
    }
    p { color: brown; }
    ol, a { font-family: serif; }
</style>

<h1>The Cat</h1>
<h2>Introduction</h2>
<p>
    My cat, a charming and playful feline, was adopted from an animal shelter and now resides in
    <strong>Seattle, WA</strong>. She has become quite the local celebrity, known not only
    for her adorable antics but also for her work as an entrepreneur.
</p>
<h2>Profile</h2>
<ol>
    <li><i>favorite food</i> - smoked salmon</li>
    <li><i>hobbies</i> - watching fish</li>
    <li><i>hidden talent</i> - Karaoke</li>
</ol>
<h2>Links</h2>
<a href="#">Seattle Animal Control Shelter</a><br>
<a href="#">Humane Society of the United States</a>
<img src="../assets/cat.jpg" alt="Cat Image" class="cat-image">
"""
display(HTML(ex08_html))
```
